```
t=[[0 for j in range(26)] for i in range(2000)]
val=[0 for i in range(2000)]
total=0

def insert(stringOne):
  u=0
  for i in range(len(stringOne)):
    id=ord(stringOne[i]) - ord('A')
    # print(id)
    if t[u][id]==0:
      global total
      total +=1
      t[u][id]= total
      
    val[t[u][id]]+=1
    print(val)
    u=t[u][id]


if __name__=="__main__":
  answer=[]
  array=[]
  inputs=int(input())
  for i in range(inputs):
    n,k= map(int,input().strip().split(" "))  
    for j in range(n):
      insert(input().strip().upper())
    ans=0
    # print(val)
    # print(total)
    for i in range(1,total+1):
      ans +=val[i]//k
    answer.append(ans)
  for i in range(inputs):
    print("Case# {}: {}".format(i+1,answer[i]))
```
